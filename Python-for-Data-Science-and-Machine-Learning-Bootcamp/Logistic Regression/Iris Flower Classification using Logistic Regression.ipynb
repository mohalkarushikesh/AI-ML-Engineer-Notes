# Iris Flower Classification using Logistic Regression

This project demonstrates the use of logistic regression for multi-class classification on the classic Iris dataset. It includes data exploration, model training, prediction, and visualization of results.

## üîç Project Overview

- **Dataset**: Iris dataset (from `sklearn.datasets`)
- **Goal**: Classify iris flowers into three species (*setosa*, *versicolor*, *virginica*) based on their sepal and petal measurements
- **Model Used**: Logistic Regression (`multi_class='ovr'`)

## üìà Workflow

1. **Data Loading and Exploration**  
   - Loaded using `load_iris()`  
   - Basic data analysis with `.info()`, `.describe()`, and `pairplot`

2. **Model Building**  
   - Split data into training and test sets (80/20)  
   - Trained a logistic regression model

3. **Evaluation**  
   - Calculated accuracy score  
   - Visualized actual vs. predicted labels using bar and line plots

````python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import seaborn as sns
# dataset
from sklearn.datasets import load_iris
# model selection 
from sklearn.model_selection import train_test_split
# model 
from sklearn.linear_model import LogisticRegression
# metrics 
from sklearn.metrics import accuracy_score
````

````python
# load the iris data
iris = load_iris()
x, y = iris.data, iris.target
````

````python
df = pd.DataFrame(iris.data)
df.head()
````

````python
df.info()
````

````python
df.describe()
````

````python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df)
plt.show()
````

````python
# split the data into tranning and testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
````

````python
# create logitics regression model
model = LogisticRegression(multi_class='ovr', max_iter=1000)
model.fit(x_train, y_train)
````

````python
# predictions
y_predict = model.predict(x_test)
````

````python
# Evaluation                                            
accuracy = accuracy_score(y_test, y_predict)
print(f"Accuracy: {accuracy}")
````

````python
comparison = pd.DataFrame({'Actual': y_test, 'Predicted':y_predict})
print(comparison)
````

````python
indices = np.arange(len(y_test))  # Create indices for x-axis

plt.bar(indices - 0.2, y_test, width=0.4, label='Actual', color='blue', alpha=0.6)
plt.bar(indices + 0.2, y_predict, width=0.4, label='Predicted', color='orange', alpha=0.6)

plt.xlabel('Sample Index')
plt.ylabel('Class Label')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt

# Use indices for the x-axis
indices = range(len(y_test))

# Plot actual values as a red line
plt.plot(indices, y_test, color='red', label='Actual', linestyle='-', marker='o')

# Plot predicted values as a green line
plt.plot(indices, y_predict, color='green', label='Predicted', linestyle='-', marker='x')

# Add title and labels
plt.title("Actual vs Predicted Values")
plt.xlabel("Sample Index")
plt.ylabel("Class Label")
plt.legend()
plt.show()
````
