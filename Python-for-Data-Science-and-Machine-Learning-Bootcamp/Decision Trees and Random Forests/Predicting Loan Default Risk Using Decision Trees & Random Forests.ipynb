```python
# Import necessary libraries for data handling and visualization
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# Load the dataset
df = pd.read_csv('loan_data.csv')
```

Exploratory Data Analysis

```python
# Compare FICO scores by credit policy (1 = meets credit policy)
plt.figure(figsize=(10, 6))
df[df['credit.policy'] == 1]['fico'].hist(bins=35, color='blue', label='Credit Policy = 1', alpha=0.6)
df[df['credit.policy'] == 0]['fico'].hist(bins=35, color='red', label = 'Credit Policy = 0', alpha=0.6)
plt.legend()
plt.xlabel('FICO SCORE')
```

```python
# Compare FICO scores by loan repayment status
plt.figure(figsize=(10, 6))
df[df['not.fully.paid'] == 1]['fico'].hist(bins=35, color='blue', label='Not Fully Paid = 1', alpha=0.6)
df[df['not.fully.paid'] == 0]['fico'].hist(bins=35, color='red', label = 'Not Fully Paid = 0', alpha=0.6)
plt.legend()
plt.xlabel('FICO SCORE')
```

```python
# Visualize loan purposes and their relationship to repayment status
plt.figure(figsize=(15, 7))
sns.countplot(x = 'purpose', hue='not.fully.paid', data=df, palette='Set1')
```

```python
# Joint distribution of FICO scores and interest rates
sns.jointplot(x = 'fico', y = 'int.rate', data=df, color='purple')
```

```python
# Linear model plots conditioned on credit policy and loan repayment
plt.figure(figsize=(11, 7))
sns.lmplot(y='int.rate', x = 'fico', data=df, hue='credit.policy', col='not.fully.paid', palette='Set1')
```

```python
# Create list of elements containing the string 'purpose', call it cat_feats
# Prepare categorical feature(s) for modeling by one-hot encoding
cat_feats = ['purpose']
```

```python
# Now use pd.get_dummies(df, columns=cat_feats, drop_first=True) to create fixed larger dataframe that has new feature columns 
# with new dummy variables set this dataframe as final_data
final_data = pd.get_dummies(df, columns=cat_feats, drop_first=True)
```

```python
# Preview new dataset with dummy variables
final_data.info()
```

```python
# Preview new dataset with dummy variables
final_data.head()
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Tranning Decision Tree Model

```python
# Import Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
# Fit the model 
dtree.fit(X_train, y_train)
```

```python
# Make predictions using the Decision Tree
pred = dtree.predict(X_test)
```

```python
# Evaluation of D Tree performance
from sklearn.metrics  import confusion_matrix, classification_report
print(classification_report(pred, y_test))
```

```python
print(confusion_matrix(pred, y_test))
```

Tranning Random Forests Model 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train a Random Forest classifier with 300 trees
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train, y_train)
```

```python
# Predictions for Random Forest model
predict = rfc.predict(X_test)
```

```python
# Evaluate Random Forest model
print(classification_report(predict, y_test))
```

```python
print(confusion_matrix(predict, y_test))
```
