```python
import pandas as pd
df = pd.read_csv("fake_reg.csv")
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scalar.fit(X_train)
```

```python
X_train_scaled = scalar.transform(X_train)
```

```python
X_test_scaled = scalar.transform(X_test)
```

```python
X_train.max()
```

```python
X_train.min()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

RLEU: Rectified linear unit i.e ramp function

```python
model = Sequential()

model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop', loss='mse')
```

```python
# x is the features
# y is tranning labels with are currosponds to tranning features points
# epoch : means you have gone the db 1 time i.e one pass over the entire dataset
model.fit(x=X_train_scaled, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test_scaled, y_test)
```

```python
model.evaluate(X_train_scaled, y_train, verbose=0)
```

```python
test_predictions = model.predict(X_test_scaled)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300, ))
```

```python
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis=1)
```
